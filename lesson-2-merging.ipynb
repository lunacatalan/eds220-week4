# Mergin Data 

We will learn to merge two data frames and make chloropleth map in geopandas. 

## Data

First dataset: 

Natural Earth's medium scale cultural boundaries data (1:50m).

Second Dataset:

List of Arctic communities and their location (Brook, 2023)
This is GeoJSON file.

## Data Preparation 

Import libraries


````python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
````

````python
# update display of pandas data frame 
pd.set_option('display.max.columns', None)
````

## Countries 

First, import Natural Earth polygons using `gpd.read_file` function:

````python
# import countries polygons 
countries = gpd.read_file('data/ne_50m_admin_0_countries/ne_50m_admin_0_countries.shp')
countries.head()
````

````python
# take a quick look; shapefiles for all countries in the world 
countries.plot()
````

````python
# update columns to lowercase
countries.columns = countries.columns.str.lower()
countries.head()
````

````python
# we have too many columns, only keep some
countries_sub = countries[['admin', 'type', 'geometry']]
countries_sub.head()
````

## Arctic Communities

Use `geopandas.read_file()` to read directly form URL

````python
# read in Arctic comm data
communities = gpd.read_file('https://cn.dataone.org/cn/v2/resolve/urn%3Auuid%3Aed7718ae-fb0d-43dd-9270-fbfe80bfc7a4')
communities.head()
````

````python
# check if the crs are the same
communities.crs == countries.crs
````

````python
# create canvas 
fig, ax = plt.subplots()

# plot countries on axis defined above
countries.plot(ax = ax)
communities.plot(ax = ax, color = 'orange')
plt.show()
````

We want to calculate the number of Arctic communities by country 

````python
# calculate teh number of Arctic comms by country

# extract number of communities by countrie as a Series 
n_comms = communities.groupby('country').count().name
````

````python
# convert the Series to a data frame
## rename 'name' column
n_comms = pd.DataFrame(n_comms).rename(columns = {'name':'n_communities'}).reset_index()
n_comms
````

````python
# elect Scandanavian data

scandi_codes = ['DK', 'NO', 'SE', 'FO', 'FI', 'IS', 'AX']

# select rows of countries that match list created above
scandi_n_comms = n_comms[n_comms.country.isin(scandi_codes)].copy()
scandi_n_comms
````

## Merge datasets

To merge 2 datasets they need to have at least 1 column in common. Currently, our datasets dont have any columns in common. 

````python
countries_sub.head(2)
````

````python
scandi_n_comms.head(2)
````

We can fix this by adding an `admin` column to `scandi_n_comms`. We want to join by name of country. 

````python
## add country names
scandi_names = ['Aland Islands',
               'Denmark',
               'Finland',
               'Faroe Islands',
               'Iceland',
               'Norway',
               'Sweden']

# add column
scandi_n_comms['admin'] = scandi_names
scandi_n_comms
````

To merge we can use `pandas.merge()` function. 

The basic syntax is:

```python
output_df = pd.merge(left_df, 
                     right_df,
                    how = type_of_join,
                    on = column_to_join)
```

where:
- `output_df` : dataframe resulting from merge
- `left_df` : dataframe on the left side
- `right_df` : dataframe on the left side
- `how` : type of join
- `on` : merge column

````python
merge_scandi_comm = pd.merge(countries_sub,
                             scandi_n_comms,
                             how = 'inner',
                             on = 'admin')
merge_scandi_comm
````

Notice that Aland Islands is nowheere in our countries dataframe like this:

````python
# check if this is anywhere in the original dataframe
'Aland Islands' in countries.values
````

Aland Islands is autonomous region of Finland 

````python
# uodate finland row to have an extra community
merge_scandi_comm = merge_scandi_comm.set_index('admin')
merge_scandi_comm
````

````python
# udpate single row
merge_scandi_comm.at['Finland', 'n_communities'] += 1
````

````python
merge_scandi_comm
````

## Chloropleth map

Good for aggregating map for regions

````python
merge_scandi_comm.plot(column = 'n_communities',
                      legend = True,
                      cmap = "BuPu")
````
